Q3 The columns C1 P1-P25, N1, CR1-28, M1-17 and I1-5 of the table represent the counts for 75 species grouped within five taxa: Chaetognatha (C), Polychaeta (P), Crustacea (CR), Mollusca (M), and Insecta (I). We're only interested in the richness, and we need to compute it as:
* `1` if the group has a value `> 0`
* `0` otherwise.

* Create a new column, call it `richness`, which represents the richness in each sample. The richness of `sample 1` should be `11`, since sample has non-null values only for the following groups: 
```
'C1''P6''P15''P16''P25''CR1''CR14''CR15''CR19''CR26''I3'
```



````r
survey_data$richness <-  rowSums(survey_data[ , 2:76] > 0)

head(survey_data)
````

Q4 Create a copy of the variable `survey_data` that does not have columns C1 P1-P25, N1, CR1-28, M1-17 and I1-5. Call this variable `survey_data_richness`


````r
colnames(survey_data)

survey_data_richness <- survey_data %>% select(Sample, week, angle1, angle2, exposure, salinity, temperature, NAP, penetrability, grainsize, humus, chalk, sorting1, Beach, richness)

head(survey_data_richness)
````

Q6. Use the `lm` function to model the richness as a function of the remaining variables but not including the variable `week`, which needs a special treatment we haven't covered yet!     


````r
lm_survey_data_richness <- subset(survey_data_richness, select = -c(Sample, week))

summary(lm(richness~angle1, lm_survey_data_richness))

summary(lm(richness~angle2, lm_survey_data_richness))

summary(lm(richness~exposure, lm_survey_data_richness))

summary(lm(richness~salinity, lm_survey_data_richness))

summary(lm(richness~temperature, lm_survey_data_richness))

summary(lm(richness~NAP, lm_survey_data_richness))

summary(lm(richness~penetrability, lm_survey_data_richness))

summary(lm(richness~grainsize, lm_survey_data_richness))

summary(lm(richness~humus, lm_survey_data_richness))

summary(lm(richness~chalk, lm_survey_data_richness))

summary(lm(richness~sorting1, lm_survey_data_richness))

summary(lm(richness~Beach, lm_survey_data_richness))
````

Q7. What do the various output of the `lm` mean? Interpret the results of your model. 

The residuals section represents the difference between the modeled values and the acutal values 
    >a small residual value inidicates the model has a good fit
    
The residuals standard error is the standard deviation of the residuals

The coefficients represent the variable I tested against the richness

The Pr value is the p-value for that comparison, and indicates significance
    
The adjusted R-squared value shows the amount of variance that is explained by the model. 

-Looking at the output from all of these models it appears as though angle2 (p =0.02), exposure (p < 0.01), salinity (p <0.01), NAP (p < 0.01), penetrability (p < 0.01), grain size (p < 0.01), humus (p < 0.01), chalk (p < 0.01), sorting1 (p < 0.01), and beach (p < 0.01) all significantly correlate with richness

- Although most of the models appear to have fairly large min and max residuals as well as large residual standard errors, indicating that potentially these models do not have a great fit.
